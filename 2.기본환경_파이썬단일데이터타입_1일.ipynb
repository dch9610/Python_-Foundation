# 파이썬-단일데이터타입

- 수치형 (정수, 부동소수), 불린형, 문자열
- 변수
  - 값을 담는 그릇
  - 단, 파이썬에서는 데이터의 실체를 가르키는 주소를 가진 그릇이다.
  => **파이썬은 모든것이 객체이다**. 객체를 표현하는 변수는 주소만 가진다.
  

```
# 객체를 확인하는 방법
# 도트 연산자 사용(.) 활용
# a(변수)를 입력하고 잠시 기다리면 인텔리전스가 작동하여 후보들이 자동으로 나타난다. (객체)
# 만약 아나콘다 주피터를 사용한다면 .을 입력후 TAB키를 누르면 나타난다.
```

```
# a = 1을 참조하고 있다가,
a = 1.1
# a는 1.1로 변경되었다
```

```
# 왜 가능한가?
# 1. 파이썬은 타입 추론형 언어 => 값이 변수에 세팅될때 타입이 결정된다.
# <-> 자바/C++/C# int a => 정수만 와야한다.
a=1 # 값이 들어갈때 결정 => 파이썬, js, 코클린, 스위프트
# 2. 파이썬은 모든 것이 객체이다. 1에서 1.1로 바뀐것은 단지 참조값만 변경됬다. (주소만 변경)
# a는 정수, 부동소수형, 문자열, 객체, 함수 다 가능
# 따라서, 사전에 정의된 클래스, 내장함수, 외장함수등 이름을 변수로 사용하면 안된다.
```
