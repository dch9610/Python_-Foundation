### 11. 포맷팅

```
# 11. 포맷팅
# 문자열과 문자열의 결합, 문자열 안에 특정 데이터를 삽입하여 생성
# 11-1, 문자열 % (파라미터)
# 11-2, 문자열.format(파라미터)
# 11-3, f'문자[파라미터]열' <- 가장 최근에 추가되었다
# 데이터를 원하는 형태로 가공 처리 위해서 사용 -> 정제의 역할
# 향후 문자열 전처리 => 정규식 처리 (단독 주제)


a = 1
b = 2

# 타입이 맞지 않아서 에러
#a + " + " + b = (a+b)
```

```
str(a) + " + " + str(b) + ' = ' + str(a+b)
```

```
# 포맷팅 => 결과를 먼저 쓰고 나머지는 채운다.
"%d + %d = %d" %(a,b,a+b)
```

```
# 11-1
# 위의 표현은 타입을 제한적으로 사용해서 타입이 틀리면 잠재적인 오류가 예상
# 범용적으로 표현하여 예방
# %s는 입력값의 타입이 뭔지 몰라도 다 처리해준다.
# 단, 순서를 바꿀수 없다, 순서대로 자리에 들어간다.

"%s + %s = %s" %(a,b,a/b)
```

```
# 11-2
# 위의 문제점등을 개선
"{} + {} = {}" .format(a,b,a/b)
```

```
# 순서를 바꿔도 가능
"{1} + {0} = {2}" .format(a,b,a/b)
```

```
# 파라미터를 사용하지 않아도 가능
"{1} + {1} = {0}" .format(a,b,a/b)
```

```
# 의미를 명확하게 부여할수 있다
"{height} + {weight} = {result}" .format(height= a,weight=b,result=a/b)
```

```
# 11-3 
# f -string 포맷팅, 가장 간결
f'{a}+{b}={a+b}'
```

### 14. 문자열 포맷팅 응용에서 값의 자리수(칸수) 지정

```
# 파라미터 값이 하나 이상이면 튜플형식으로
'[%s]' % 'abcd'
```

```
# 10칸을 무조건 차지한다. 그 외는 공백으로 처리
# 양수는 왼쪽 공백, 음수는 오른쪽 공백
'[%10s]' % 'abcd', '[%-10s]' % 'abcd'
```

```
'[%10f]' % 3.145678, '[%10s]' % 3.145678
```

```
'[%10.2f]' % 3.145678
```

```
# format 함수를 사용하여 공백처리
'[{:10}]'.format('abcd')
```

```
#'[{:-10}]'.format('abcd') : 역방향은 에러 발생
# < 왼쪽으로 공백으로, > 오른쪽을 공백으로, ^ 값을 중앙으로 
'[{:<10}]'.format('abcd'), '[{:>10}]'.format('abcd'),'[{:^10}]'.format('abcd')
```

```
# 공백을 *으로 대체
'[{:*^10}]'.format('abcd')
```

```
a = 'abcd'
f'[{a:10}]', f'[{a:>10}]' , f'[{a:*^10}]'
```
