# if, elif, else

- 만약 xx라면, xx해라
- 만약 xx이면, xx하고, xx이면 xx하고..
- 가정이 들어간다. 케이스가 존재한다.


```
# 현재 커피숍이다. 커피를 구매하고 결제를 하려는 상황이다.
# 직원이 물어본다. 1500원입니다.
# input을 통해서 입력받은 데이터는 무조건 문자열이다.

coffee_price = input("구매하신 커피는 1500원입니다.")
coffee_price, type(coffee_price)
```

```
# 문자열 -> 수치로
# 단, 수치로 될수 없는 표시들을 제거하거나 다시 입력 받아야함
coffee_price = int(coffee_price)
type(coffee_price)
```

```
# 포스에서 계산, 케이스는 현금결제모드 3개 (과하게, 딱맞게, 부족하게 입금)
# 조건에 맞춰서 분기해야한다.

if 조건:
  # pass는 포지션을 차지하는데, 아무일도 하지 않는다
  # 조건문, 반복문, 함수, 클래스등 틀을 잡을때 사용
  pass
  print("")
elif 조건: 
  pass
  print("")
else:
  pass
```

```
# 상수
COFFEE_ICE_PRICE  = 1500
```

```
# 돈을 더 많이 주었을 경우
# 비교 연산자 : >, <, >=, <=, ==, !=
# 비트 연산자 : not, and, or
coffee_price = int(input("구매하신 커피는 1500원입니다."))

# 퍼포먼스를 고려하여 가장 많은 케이스를 먼저 표시
# 돈을 더 많이 주었을 경우
if coffee_price > COFFEE_ICE_PRICE:
  print(f'거스름돈이 {coffee_price-COFFEE_ICE_PRICE}원 입니다.')
  pass
# 같은 가격을 입력받았다면  
elif  coffee_price == COFFEE_ICE_PRICE: 
  print("안녕히 가세요!")
  pass
# 부족하게 냈다.  
else: 
  print(f'부족한 금액은 {COFFEE_ICE_PRICE-coffee_price}원 입니다.')
  pass
```

### 부정상황 잡기

# for

- for ~ index 방식, for ~ each 방식
- 파이썬은 for ~ each만 지원한다

- 특정한(지정된) 횟구만큼 반복한다
  - for
- 언제 끝날지 모르겠는데? (0~무한대)
  - while

```
# 연속데이터 타입의 변수명은 복수로 묘사한다.
items = [1,2,3,4,5]

# 한개씩 꺼내서 출력하시오
# items 안에서 item을 한개씩 순서대로 뽑아낸다.
for item in items:
  print(item)
```

```
items = [(1,2,7),(3,4,8),(5,6,9)]
print(len(items))

for item in items:
  print(item)

# 값들을 직접 변수로 받아서 데이터를 획득하고 싶다.

for i in range(0,len(items)):
  for j in range(0,len(items[i])):
    print(items[i][j])
```

```
# 환율 데이터
exchanges = [
    {
        'code':'USD',
        'buy':1118.11
    },
    {
        'code':'EUR',
        'buy':1318
    }
]
# 리스트에 dic가 포함되어있는 형태
exchanges[0]
```

```
# 데이터를 돌면서
# ex) "USD를 구매할때 환화로 1318원이 필요합니다." 이렇게 출력

for exchange in exchanges:
  # 딕셔너리에서 키를 넣어서 값을 추출 => 인덱싱
  # 인덱싱 => 변수[ 키 ]
  # 문자열이나 코드가 길어져서 내려친다면 \ 표현하여 이어짐 표시
  print(f"{exchange['code']}를 구매할때,\
     환화로 {exchange['buy']}원이 필요합니다.")
  
```

```
# 딕셔너리 데이터를 for문으로 사용해본다면
# 키값이 뽑힌다.
for key in exchanges[0]:
  print(key)
```

```
tmp = exchanges[0]

for key in exchanges[0]:
  # 값을 출력하고 싶다면
  print(exchanges[0][key])
```

```
# .items() : [(키,값), (키,값), ...]
tmp.items()
```

```
# 딕셔너리 값을 좀 더 심플하게 키와 값을 추출해본다
# dic 함수 .items를 사용하면 키와 값을 동시에 추출할수 있다
for k,v in tmp.items():
  print(k, v)
```

### enumerate 사용

```
# 딕셔너리는 순서가 없기 때문에 키가 나오는 순서대로 
# 임의로 인덱스를 붙인다 (0, 1, 2, ...)
for index, key in enumerate(tmp):
  print(index, key, tmp[key])
  #print(key, tmp[key])
```

### range 사용

```
# 연속된 수를 나열하는 range()
# x <= n < y : stop 부분은 포함되지 않는다.
# 기본 step(값이 증가하는 간격)은 1이다.
# for n in range(1, 10) : 1 ~ 9
# for n in range(10) : 0 ~ 9 
for n in range(1,10):
  print(n)
```

```
for n in range(0,10,2):
  print(n)
```

```
# range()를 활용하여 비트연산 (0 or 1)
# 0, 0, bit 연산 -> and, or, xor : 출력값
# 2중 for문 케이스 => 모든 케이스를 만들때
for n in range(2): # 0,1
  for m in range(2): # 0,1
    print(f'n={n}, m={m} bit연산 -> {n and m}, {n or m},\
    {n ^ m} ')
```

### 구구단 출력

```
# 구구단 3단 ~ 5단까지 구현하시오
# 출력내용 3 x 1 = 3, ... 5 x 9 =45 

for n in range(3,6):
  for m in range(1,10):
    print(f'{n}x{m}={n*m:>2}') # 2자리 정렬
  print("="*20)
```

```
# Lv2. 3단, 5단만 출력하시오
for n in range(3,6,2):
  for m in range(1,10):
    print(f'{n}x{m}={n*m:>2}')
  print("="*20)
```

```
# 결과는 list에 모두 담는다
# 1. 담을 그릇을 준비 => 리스트

results = []

for n in range(3,6,2):
  for m in range(1,10):
    # 2. 담는다 -> 리스트에 멤버로 추가한다.
    results.append((f'{n}x{m}={n*m:>2}'))
    print(f'{n}x{m}={n*m:>2}')
  print("="*20)
# 3. 확인
len(results)
```
